## Задание 2: Уникальные символы в тексте

```python
text_file = 'task2_text.txt'
russian_text = """
Вот такой текст.
Здесь есть разные буквы и символы.
Некоторые буквы повторяются, а некоторые - нет.
Посмотрим, что получится в результате.
"""

module2.create_text_file(text_file, russian_text)
print("Текст создан:")
print(russian_text)
```

```python
unique_chars = module2.find_unique_chars(text_file)
print(f"Символы, встречающиеся 1 раз: '{unique_chars}'")
print(f"Количество уникальных символов: {len(unique_chars)}")
```

```python
with open(text_file, 'r', encoding='utf-8') as f:
    text = f.read()

freq = Counter(text)

print("Частота всех символов:")
for char, count in sorted(freq.items()):
    if char == '\n':
        print(f"Символ '\\n' (перевод строки): {count}")
    elif char == ' ':
        print(f"Символ 'пробел': {count}")
    elif char == '.':
        print(f"Символ 'точка': {count}")
    elif char == ',':
        print(f"Символ 'запятая': {count}")
    else:
        print(f"Символ '{char}': {count}")
```

```python
print("Уникальные символы в порядке появления:")
unique_list = []
for i, char in enumerate(text):
    if text.count(char) == 1:
        unique_list.append(char)
        if char == '\n':
            print(f"  {len(unique_list)}. '\\n' (перевод строки)")
        elif char == ' ':
            print(f"  {len(unique_list)}. 'пробел'")
        else:
            print(f"  {len(unique_list)}. '{char}'")
```

```python
test_text = "Мама мыла раму"
test_file = 'test_text.txt'
module2.create_text_file(test_file, test_text)

test_unique = module2.find_unique_chars(test_file)
print(f"Текст: '{test_text}'")
print(f"Уникальные символы: '{test_unique}'")

print("\nПроверка вручную:")
for char in test_text:
    if test_text.count(char) == 1:
        print(f"  '{char}' встречается 1 раз")
    else:
        print(f"  '{char}' встречается {test_text.count(char)} раза")
```
